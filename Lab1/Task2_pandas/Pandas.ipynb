### 7) Використання функції loc. Виведемо запис, де значення у колонці 'country-code' дорівнює 16.

### 8) Використання функції fillna(), яка дозволяє заповнити пропуски в даних зі значенням Nan на будь-яке значення. Замінимо Nan на 0.

```python
print("Початкові дані: ")
url = 'https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv'
data_country = pd.read_csv(url)
data_country
```

```python
print("Оновлені дані: ")
data_country = data_country.fillna(0)
data_country
```
